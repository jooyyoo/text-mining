**그리디 서치 디코딩**

기계 번역 및 자연어 생성에서 가장 일반적인 디코딩 방법 중 하나이다.

이 방법은 각 시점에서 가장 가능성이 높은 단어를 선택하여 문장을 생성한다.

그리디 서치 디코딩은 각 시점에서 가능한 모든 단어의 조합을 고려하지 않고, ***단순히 현재 상태에서 가장 가능성이 높은 단어를 선택한다***

그리디 서치 디코딩은 간단하고 빠르지만, 최적의 결과를 보장하지 않는다. 이는 선택된 단어가 이후 시점에서의 최적의 단어 조합을 방해할 수 있기 때문이다. 

***빔서치***와 같은 다른 디코딩 방법들은 가능한 모든 단어의 조합을 고려하여 최적의 결과를 찾아내지만, 계산 비용이 높아진다는 단점이 있다.

```python
import matplotlib.pyplot as plt
import numpy as np

def softmax(logits, T=1):
  e_x = np.exp(logits/T)
  return e_x / e_x.sum()

logits = np.exp(np.random.random(1000))
sorted_logits = np.sort(logits)[::-1]
x = np.arange(1000)

for T in [0.5,1.0,2.0]:
  plt.step(x,softmax(sorted_logits,T), label=f"T={T}")


plt.legend(loc="best")
plt.xlabel("sorted token probabilites")
plt.ylabel("probobility")
plt.show()
```

T 값이 작아진다
*  1등이 점수를 많이 받게 된다. 확률값이 높은 아이들을 보고 싶다면,,,

T 값이 커진다
* 등수 낮은 애들도 가능성을 가진다
* 하위 등수도 뽑힐 가능성이 있다는 것
* 골고루 뽑힐 수 있게 됨
